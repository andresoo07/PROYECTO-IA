```python
#  Exploración inicial

# 1. Importamos librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuramos estilo de gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 2. Cargamos el archivo CSV
df = pd.read_csv("/train.csv")

# 3. Ver las primeras filas para inspección inicial
print("Primeras filas del dataset:")
display(df.head())

# 4. Información general del dataset
print("\nInformación general del dataset:")
df.info()

# 5. Estadísticas descriptivas de columnas numéricas
print("\nEstadísticas descriptivas:")
display(df.describe())

# 6. Distribución de clases (RENDIMIENTO_GLOBAL)
print("\nDistribución de clases (RENDIMIENTO_GLOBAL):")
print(df["RENDIMIENTO_GLOBAL"].value_counts())

# Gráfico de barras de la variable objetivo
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="RENDIMIENTO_GLOBAL", order=df["RENDIMIENTO_GLOBAL"].value_counts().index)
plt.title("Distribución de RENDIMIENTO_GLOBAL")
plt.xlabel("Nivel de desempeño")
plt.ylabel("Cantidad de estudiantes")
plt.show()

# 7. Top 10 departamentos con más estudiantes
if "ESTU_PRGM_DEPARTAMENTO" in df.columns:
    print("\nTop 10 departamentos con más estudiantes:")
    top_dept = df["ESTU_PRGM_DEPARTAMENTO"].value_counts().head(10)
    print(top_dept)

    plt.figure(figsize=(10, 6))
    sns.barplot(y=top_dept.index, x=top_dept.values, hue=top_dept.index, legend=False)
    plt.title("Top 10 departamentos con más estudiantes")
    plt.xlabel("Cantidad de estudiantes")
    plt.ylabel("Departamento")
    plt.show()
else:
    print("No se encontró la columna ESTU_PRGM_DEPARTAMENTO.")
```
